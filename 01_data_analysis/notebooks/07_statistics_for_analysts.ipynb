## Objective
Understand statistical concepts from a business perspective and learn how to interpret data correctly for decision-making rather than memorizing formulas.

```python
import pandas as pd
df = pd.read_csv("../data/retail_sales_cleaned.csv")
```

```python
df["revenue"] = df["quantity"]*df["unit_price"]
```

```python
mean_revenue = df["revenue"].mean()
```

```python
median_revenue = df["revenue"].median()
```

```python
mean_revenue, median_revenue
```

# Difference in Mean and Median

## Mean - affected by outliers (extreme values)
## Median - resistant to outliers

By analogy we can say - If billionares increases, the mean will increase but middle class men will not be affected by that.

```python
# How to know if a business is stable or not?
```

```python
# Low std deviation means low volatility - predictable business
# High std deviation means high volatility - risky business
```

```python
var = df["revenue"].var()
std_dev = df["revenue"].std()
```

```python
var, std_dev
```
