# Industry Project 1: End-to-End Sales Analytics

## Objective
Build a complete analytics pipeline including data cleaning, KPI definition, SQL modeling, and dashboard-ready datasets for executive reporting.


```python
import pandas as pd

df = pd.read_csv("../../../data/retail_sales_cleaned.csv")
df["order_date"] = pd.to_datetime(df["order_date"])
df["revenue"] = df["quantity"] * df["unit_price"]
```

```python
df
```

```python
import os
os.getcwd()
```

```python
total_revenue = df["revenue"].sum()
avg_order_val = df["revenue"].mean()
total_orders = df["order_id"].nunique()
top_category = df.groupby("category")["revenue"].sum().idxmax()
top_region = df.groupby("region")["revenue"].sum().idxmax()

total_revenue, avg_order_val, top_category, top_region, total_orders
```

```python
df.to_csv("../data/sales_dashboard_dataset.csv", index = False)
```

```python
revenue_by_category = df.groupby("category")["revenue"].sum().reset_index()
revenue_by_category
```

```python
revenue_by_region = df.groupby("region")["revenue"].sum().reset_index()

revenue_by_region
```

```python
monthly_revenue = df.groupby(df["order_date"].dt.to_period("M"))["revenue"].sum().reset_index()

monthly_revenue["order_date"] = monthly_revenue["order_date"].astype(str)

monthly_revenue
```

```python
top_products = df.groupby("product")["revenue"].sum().reset_index()

top_products = top_products.sort_values(by="revenue", ascending=False).head(10)

top_products
```
