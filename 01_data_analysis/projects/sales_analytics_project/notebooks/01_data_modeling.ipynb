## Objective
Build a complete analytics pipeline including data cleaning, KPI definition, SQL modeling, and dashboard-ready datasets for executive reporting.


```python
import pandas as pd

df = pd.read_csv("../../../data/retail_sales_cleaned.csv")
df["order_date"] = pd.to_datetime(df["order_date"])
df["revenue"] = df["quantity"] * df["unit_price"]
```

```python
total_revenue = df["revenue"].sum()
avg_order_val = df["revenue"].mean()
top_category = df.groupby("category")["revenue"].sum().idxmax()
top_region = df.groupby("region")["revenue"].sum().idxmax()

total_revenue, avg_order_val, top_category, top_region
```
